```python
# prompt: import pandas,numpy,matplt and seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/Vivekjangir90/YBI_ML_PROJECTS/raw/main/train.csv")
df
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',10)
```

```python
df.head(10)
```

```python
null_var = df.isnull().sum()/df.shape[0]*100
null_var
```

```python
drop_col = null_var[null_var>20].keys()
drop_col
```

```python
df_drop_col = df.drop(columns=drop_col)
df_drop_col.shape
```

```python
df_num = df_drop_col.select_dtypes(include=['int64','float64'])
df_num.shape
```

```python
a=df_num[df_num.isnull().any(axis=1)]
a
```

```python
df_num.isnull().sum()
```

```python
missing_num_var = [var for var in df_num.columns if df_num[var].isnull().sum()>0]
missing_num_var
```

```python
df_num_mean = df_num.fillna(df_num.mean())
df_num_mean.isnull().sum()
```

```python
sns.heatmap(df_num_mean.isnull())
```
