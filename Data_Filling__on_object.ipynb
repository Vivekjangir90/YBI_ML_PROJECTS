```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/Vivekjangir90/YBI_ML_PROJECTS/raw/main/train.csv")
df.head()
```

```python
cat_vars=df.select_dtypes(include='object')
cat_vars.head()
```

```python
missing_value = cat_vars.isnull().mean()*100
missing_value
```

```python
drop_var = [var for var in missing_value.index if missing_value[var]>30]
drop_var
```

```python
drop_col_df = cat_vars.drop(columns=drop_var)
```

```python
isnull_per = drop_col_df.isnull().mean()*100
miss_vars = isnull_per[isnull_per > 0].keys()
miss_vars
```

```python
drop_var = missing_value[missing_value > 30].keys()
drop_var
```

```python
cat_vars['MasVnrType'].mode()
```

```python
cat_vars['MasVnrType'].value_counts()
```

```python
cat_vars['MasVnrType'].fillna(cat_vars['MasVnrType'].mode()[0])
```

```python
for var in miss_vars:
  cat_vars[var].fillna(cat_vars[var].mode()[0],inplace=True)
  print(var,"=",cat_vars[var].mode()[0])
```
